```python
games = games[games.site != 'Neutral']
big_dance= big_dance[big_dance["Year"]>2012]
```

```python
games = games[games["date"]>'2012-07-01']

season_list=[]
for index,row in games.iterrows():
    year=int(row[2][0:4])
    month=int(row[2][5:7])
    day=int(row[2][7:])
    if (month-1)*30+day>180:
        season_list.append(int(row[2][0:4])+1)
    else:
        season_list.append(int(row[2][0:4]))

games['Season']=season_list
```

```python
#Loading current bracket predictions to MongoDB
mconn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(mconn)
db = client.bracket_db
collection = db.team
```

```python
url = 'http://www.espn.com/mens-college-basketball/bracketology'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = BeautifulSoup(response.text, 'lxml')
```

```python
bracket_team = soup.find_all('div', class_='team')
bracket_list = []
for bracket in bracket_team:
    team = {'Team Name':bracket.a.text,
           'Date':datetime.datetime.now().strftime('%Y-%m-%d')}
    collection.insert_one(team)
    bracket_list.append(bracket.a.text)
```
