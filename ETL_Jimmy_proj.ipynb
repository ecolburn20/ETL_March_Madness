```python
games = games[games.site != 'Neutral']
games.head(100)
```

```python
games = games[games["date"]>'2012-07-01']
```

```python
games.head(100)
```

```python
season_list=[]
for index,row in games.iterrows():
    year=int(row[2][0:4])
    month=int(row[2][5:7])
    day=int(row[2][7:])
    if (month-1)*30+day>180:
        season_list.append(int(row[2][0:4])+1)
    else:
        season_list.append(int(row[2][0:4]))

games['Season']=season_list
```

```python
games=games.rename(index=str, columns={"team_name": "Team", "game_num": "Game Number","date":"Date","opp_name":"Opponent","w_l":"Win or Loss","pts":"Points","opp_pts":"Opponent Points","fg":"Field Goals","fga":"Field Goal Attempts","fg_per":"Field Goal Percentage","3p":"Three Pointers","3p_per":"Three Point Percentage","ft":"Free Throws","fta":"Free Throw Attempts","ft_per":"Free Throw Percentage","orb":"Offensive Rebounds","trb":"Total Rebounds","ast":"Assists","stl":"Steals","blk":"Blocks","tov":"Turnovers","pf":"Personal Fouls","opp_fg":"Opponent Field goals","opp_fga":"Opponent Field Goal Attempts","opp_fg_per":"Opponent Field Goal Percentage","opp_3p":"Opponent Three Points","opp_3pa":"Opponent Three Point Attempts","opp_3p_per":"Opponent Three Point Percentage","opp_ft":"Opponent Free Throw","opp_fta":"Opponent Free Throw Attempts","opp_ft_per":"Opponent Free Throw Percentage","opp_orb":"Opponent Offensive Rebounds","opp_trb":"Opponent Total Rebounds","opp_ast":"Opponent Assists","opp_stl":"Opponent Steals","opp_blk":"Opponent Blocks","opp_tov":"Opponent Turnovers","opp_pf":"Opponent Personal Fouls"})
```

```python
games.head()
```

```python
###Analysis time###
wins_count=[]
for index,row in games.iterrows():
    if row[5]=="W":
        wins_count.append(1)
    else:
        wins_count.append(0)

games["wins_help"]=wins_count
game_wins=games.groupby(by=["Team","Season"])
```

```python
game_help=game_wins.sum()
array_help=games["Team"]
game_help["team help"]=array_help
#array_help[0]
```

```python
game_help=game_help.reset_index()
```

```python
game_help.head(100)
```

```python
analysis_tuple=[]
for index,row in game_help.iterrows():
    help=(row[0],row[2],row[-2])
    analysis_tuple.append(help)
    

analysis_tuple
```

```python
#lets pick the 5 best teams
final_teams=["Duke","Virginia","Gonzaga","North Carolina","Michigan"]
tournament_wins_count=[]
for x in range(len(final_teams)):
    tournament_wins_count.append(0)
    for index,row in big_dance.iterrows():
        if final_teams[x]==row[6] and row[5]>row[8]:
            tournament_wins_count[x]+=1
        if final_teams[x]==row[7] and row[8]>row[5]:
            tournament_wins_count[x]+=1
tournament_wins_count
```

```python
help_teams=["Duke Blue Devils","Virginia Cavaliers","Gonzaga Bulldogs","North Carolina Tar Heels","Michigan Wolverines"]
regular_season_wins=[]
for t in range(len(help_teams)):
    regular_season_wins.append(0)
    for i in range(len(analysis_tuple)):
        if analysis_tuple[i][0]==help_teams[t]:
            regular_season_wins[t]+=analysis_tuple[i][2]
regular_season_wins
```

```python
#plotting the teams to see the correlation between regular season wins and tournament wins
plt.scatter(regular_season_wins,tournament_wins_count)
```
